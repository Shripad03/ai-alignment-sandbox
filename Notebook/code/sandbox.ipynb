```python
print("Missing value per column")
print(df.isnull().sum())
```

```python
for col in df.columns :
    if df[col].dtype == "object" :
        df[col] = df[col].fillna("Unknown")
    else :
        df[col] = df[col].fillna(df[col].mean())
```
